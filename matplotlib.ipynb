# matplotlib Demo
matplotlib is a versatile plotting library for Python that can generate a wide variety of charts and visualizations.

Important:
Always remember to integrate `%matplotlib inline` at the beginning of your Jupyter Notebook to display plots directly in the notebook.

Import matplotlib. (Also import numpy, which is needed for an example):

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Line Plot

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.show()
```

## Bar Chart

```python
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

plt.bar(categories, values, color=['red', 'green', 'blue', 'cyan'])
plt.title("Simple Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

## Scatter Plot

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y, color='purple', marker='o')
plt.title("Simple Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

## Histogram

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Simple Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

## Pie Chart

```python
labels = ['A', 'B', 'C', 'D']
sizes = [3, 4, 6, 2]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.2, 0, 0, 0)  # explode the first slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures the pie is circular.
plt.title("Simple Pie Chart")
plt.show()
```

## Subplots

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))

# Plotting on the individual subplots
axes[0, 0].plot(x, y, 'r')
axes[0, 0].set_title('Line Plot')
axes[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'cyan'])
axes[0, 1].set_title('Bar Chart')
axes[1, 0].scatter(x, y, color='purple', marker='o')
axes[1, 0].set_title('Scatter Plot')
axes[1, 1].hist(data, bins=30, color='skyblue', edgecolor='black')
axes[1, 1].set_title('Histogram')

plt.tight_layout()
plt.show()
```
